THE METRIX

The metric is the numeric quantity on which we rank movies.
- we need a metrix which can to an extent take into account the movie rating and the number of votes(popularity).
- This would give a greater preference to a blockbuster movie rated 8 by 1,00,000 user over an art movie rated 9 by 100 users.
- Movies with few ratings are not that much reliable.
- A movie rated 10/10 by 5 users does not necessarily means that is a good movie.
- Here we are using IMBD's weighted rating formula
       
   **ur = (( v / v + m ) * R) + (( m / v + m ) * C)**
        
      here, 
         
  -  v : is the number of votes garnered by the movie
  -  m : is the minimum number of votes required for the movie to be in the chart(the prerequisite)
   - R : mean rating of the movie
   - C : mean rating of all the movies in the dataset

### THE PREREQUISTIES

- There is no fixed value for m(it is an arbitary value)
- For our purpose we are using 80th percentile movie as our value for m.
- For a movie to be considered in the rankings,it must have garnered more votes that at least 80% of the movies present in our dataset

```python
# calculate the number of votes garnered by the 80th percentile movie
m = movies["vote_count"].quantile(0.80)
m
```

### consider movies only more than 45 minutes and less than 300 minutes

```python
movies["runtime"].dtype
```

```python
# movies <45 and <300
q_movies = movies[(movies["runtime"] > 45 ) & ( movies["runtime"] <=300)]
```

```python
q_movies
```

```python
q_movies = q_movies[q_movies["vote_count"] >= m]
```

```python
q_movies.shape
```

```python
movies.shape
```

### Calculate the score

```python
C = movies["vote_average"].mean()
```

```python
C
```

```python
# function to calculate the IMBD weighted rating for each movie
def weighted_rating(x,m=m,C=C):
    v = x["vote_count"]
    R = x['vote_average']
    # compute the weighted rating
    return (v/(v+m)*R) + (m/(m+v)*C)
```

```python
q_movies["score"] = q_movies.apply(weighted_rating,axis=1)
```
